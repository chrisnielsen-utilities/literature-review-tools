### Create API client

### Pull data from Openreview API

```python
conference_url = 'NeurIPS.cc/2021/Conference'
```

```python
submissions = openreview.tools.iterget_notes(client, invitation= conference_url + '/-/Blind_Submission')
submissions_by_forum = {n.forum: n for n in submissions}
```

```python
comments_list = []
for i, x in enumerate(submissions_by_forum):
    print(i, len(submissions_by_forum))
    
    notes = submissions_by_forum[x]
    forum_comments = client.get_notes(forum=notes.forum)
    comments_list.append( (notes,forum_comments) )
    
```

### Save raw data

```python

with open('data/2021_neurips_raw_paper_data.pkl','wb') as file:
    pickle.dump(comments_list, file)
```

```python
with open('data/2021_neurips_raw_paper_data.pkl','rb') as file:
    comments_list = pickle.load(file)
```

### Extract important fields from data

```python
paper_data = []

for paper in comments_list:
    
    paper_notes = paper[0]
    paper_metadata_list = paper[1]
    
    url = paper_notes.content['_bibtex'].split('url={')[1][:-3]
    pdf = 'https://openreview.net' + paper_notes.content['pdf']
    title = paper_notes.content['title']
    
    if 'code' in paper_notes.content:
        code = paper_notes.content['code']
    else:
        code = None
    keywords = paper_notes.content['keywords']
    
    if 'TL;DR' in paper_notes.content:
        tldr = paper_notes.content['TL;DR']
    else:
        tldr = None
        
    abstract = paper_notes.content['abstract']
    
    venue = paper_notes.content['venue']
    
    time_spent_reviewing_list = []
    rating_list = []
    confidence_list = []
    
    for paper_metadata in paper_metadata_list:
        for reply in paper_metadata.details['directReplies']:
            try:
                time_spent_reviewing = reply['content']['time_spent_reviewing']
                rating = reply['content']['rating']
                confidence = reply['content']['confidence']

                time_spent_reviewing_list.append(time_spent_reviewing)
                rating_list.append(rating)
                confidence_list.append(confidence)
            except:
                pass

    
    paper_data_dict = {}
    
    paper_data_dict['url'] = url
    paper_data_dict['tldr'] = tldr
    paper_data_dict['pdf'] = pdf
    paper_data_dict['time_spent_reviewing_list'] = time_spent_reviewing_list
    paper_data_dict['rating_list'] = rating_list
    paper_data_dict['confidence_list'] = confidence_list
    paper_data_dict['title'] = title
    paper_data_dict['code'] = code
    paper_data_dict['keywords'] = keywords
    paper_data_dict['abstract'] = abstract
    paper_data_dict['venue'] = venue
    
    paper_data.append(paper_data_dict)
            
    
  
    
```

### Save results

```python
with open('data/2021_neurips_processed_paper_data.pkl','wb') as file:
    pickle.dump(paper_data, file)
```
